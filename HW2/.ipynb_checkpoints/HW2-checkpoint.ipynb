## Fixing the Correlations

It occurs to me after carefully observing the above plots that the data may be misleading.  The reason I came to this conclusion is that correlation measures how much two variables have a linear relationship, however the logerror inherently has negative values when there is a negative error.

To attempt to remedy this situation let us take the Pearson and Spearman correlations applied to the absolute value of the log error.

```python
abs_training = training.copy()
abs_training['logerror'] = abs_training['logerror'].abs()

# Pearson Correlation
abs_error_correlation = abs_training.corr(method='pearson')['logerror'][:-1].select(lambda x: not re.search('id', x))
y_pos = np.arange(len(abs_error_correlation.values))
plt.barh(y_pos, abs_error_correlation.values, align='center', alpha=0.5)
plt.yticks(y_pos, abs_error_correlation.index)
plt.xlabel('Pearson Correlation')
plt.title('Correlations between features and log error')
plt.show()
```

```python
abs_s_error_correlation = abs_training.corr(method='spearman')['logerror'][:-1].select(lambda x: not re.search('id', x))

# Spearman Correlation
y_pos = np.arange(len(abs_s_error_correlation.values))
plt.barh(y_pos, abs_s_error_correlation.values, align='center', alpha=0.5)
plt.yticks(y_pos, abs_s_error_correlation.index)
plt.xlabel('Spearman Correlation')
plt.title('Correlations between features and log error')
plt.show()
```

After correcting for the absolute value, we see that the two correlation graphs have changed significantly.  There are more negative correlations in the Pearson graph and they are more pronounced.  The entire shape of the Spearman graph has transformed and we can easily see which variables have the most influence on the log error.

Unfortunately, although we can now better localize interesting features, there still is not a particularly large correlation between any single variable and the log error if we consider the $r^2$ metric.

For completeness we now investigate the scatter plot with the absolute value of the log error and the basementsqft as before.

```python
plt.scatter(abs_training['basementsqft'], abs_training['logerror'])
plt.ylabel('Log Error')
plt.xlabel('Basement sqft')
plt.show()
```

Again, it seems that the Pearson correlation is being heavily affected by the outliers present in the relationship.

# Simple Regression

Having thoroughly examined our dataset and the relationships present in the features we are in the position to run a rudimentary regression.  In a way this regression will serve as yet another exploration of the space we are working in.

To begin with, we will need to partition the training set into a train/test split to evaluate how well the learner works.

Next we will need to decide which variables to run the regression on. Analyzing the corrected Pearson and Spearman plots above we can (somewhat arbitrarily) select the most correlated variables.

From inspection the following variables seem to have the most correlation (positive or negative) with the log error:

- yearbuilt
- yardbuildingsqft26
- unitcnt
- finishedsquarefeet6

Thus we will train a multivariate linear regression using these four features to predict the log error.

We will need to clean the data at this point to run it in a learning algorithm.  I will use the method of median replacement to fill in the missing data.  After filling in the missing data we will apply z-score nomalization to center the data.

```python
# Median replacement
cleaned_training = training.fillna(training.median()).select_dtypes(exclude=[object])

# zscore normalization
numeric_features = cleaned_training.select_dtypes(exclude=[object]).columns
cleaned_training[numeric_features].apply(zscore)

training_features = cleaned_training[['yearbuilt', 'yardbuildingsqft26', 'unitcnt', 'finishedsquarefeet6', 'logerror']]

# Split data into training and test sets
train, test = sklearn.model_selection.train_test_split(training_features, test_size=0.2)
```

```python
# Select the features for training
x_train = train[['yearbuilt', 'yardbuildingsqft26', 'unitcnt', 'finishedsquarefeet6']]
y_train = train['logerror']

# Select the features for testing
x_test = test[['yearbuilt', 'yardbuildingsqft26', 'unitcnt', 'finishedsquarefeet6']]
y_test = test['logerror']
```

```python
# Train a multivariate linear regression model
ols = linear_model.LinearRegression()
model = ols.fit(x_train, y_train)
```

Now that we have trained our regression we can evaluate the results on the test set using the root mean squared error metric.  This metric is suitable as it will give a good measurement of the error in our model and it is directly interpretable as it retains the same units as the original logerror.

```python
print math.sqrt(mean_squared_error(model.predict(x_test), y_test))
```

We see that our model does not do a great job at predicting the data.  Since the log error has a somewhat normal distribution centered at 0, we see that our measurements are off by a decent amount.  In fact, our model is off by almost an entire standard deviation on average.

Next we should check another metric of performance for regression, $r^2$.

```python
print r2_score(model.predict(x_test), y_test)
```

The highly negative score from $r^2$ indicates that the current simple regression performs very poorly compared to the shape of the data.  This indicates that we will need to take a stronger approach using more features and possibly a different learning algorithm.

```python
print ols.coef_
```

Finally, analyzing the coefficients of the regression we see that the third variable, unitcnt, has the most influence on the model.  This is due to the fact that it has the largest absolute valued coefficient.

## Analysis of Simple Regression

It seems clear from the above exploration that we will need to consider a much stronger model for prediction.  Some first thoughts include using more features, creating new features, using different algorithms, and mapping some features to a nonlinear function.

# A More Advanced Model

As a first exploration, let us see how well a simple multivariate regression works when using all non-string data.

```python
# Remove ID based features
cleaned_training = cleaned_training.select(lambda x: not re.search('id', x), axis=1)

# Split data into training and test sets
train, test = sklearn.model_selection.train_test_split(cleaned_training, test_size=0.2)

x_train = train.drop(['logerror'], axis=1)
y_train = train['logerror']
x_test = test.drop(['logerror'], axis=1)
y_test = test['logerror']

# Linear regression
model = ols.fit(x_train, y_train)
print 'RMSE: ' + str(math.sqrt(mean_squared_error(model.predict(x_test), y_test)))
print 'R2: ' + str(r2_score(model.predict(x_test), y_test))
```

It seems that using all of the features has greatly improved the model accuracy, unsurprisingly.  The negative $r^2$ score still implies that our model is not performing well however.  We will need to push onward.

## Comparing Algorithms

Given that the convinient implementations provided in Python let us investigate how well a variety of models work on the data.  In particular we will consider:
- Ridge Regression
- LASSO

Which both constrain the solution space of linear regression via regularization.

We will also consider k Nearest Neighbors.

### Ridge Regression

```python
ridge = linear_model.Ridge(alpha=.5)
ridge.fit(x_train, y_train)

print 'RMSE: ' + str(math.sqrt(mean_squared_error(ridge.predict(x_test), y_test)))
print 'R2: ' + str(r2_score(ridge.predict(x_test), y_test))
```

### LASSO

```python
lasso = linear_model.Lasso()
lasso.fit(x_train, y_train)

print 'RMSE: ' + str(math.sqrt(mean_squared_error(lasso.predict(x_test), y_test)))
print 'R2: ' + str(r2_score(lasso.predict(x_test), y_test))
```

Unfortunately it does not seem to be the case that either regularization helps much naively.

### k Nearest Neighbors

```python
knn = KNeighborsRegressor(n_neighbors=2)
knn.fit(x_train, y_train)

print 'RMSE: ' + str(math.sqrt(mean_squared_error(knn.predict(x_test), y_test)))
print 'R2: ' + str(r2_score(knn.predict(x_test), y_test))
```

It seems that a nearest neighbor approach is performing much better than simple regression with respect to $r^2$.

## Princial Components Analysis

Since we are working with many features I would like to try and reduce the dimensionality of the space by using Principal Components Analysis (PCA).

We will tune the number of dimensions to project onto via iterative search over the component parameter.

```python
r_scores = []  # Holds linear regression r^2 scores
rmse_scores = []  # Holds RMSE scores

ols = linear_model.LinearRegression()

# Iterate over component sizes, recording the r^2 score for each model via linear regression
for i in range(1, len(x_train.keys())):
    
    # Run PCA
    pca = PCA(n_components = i)
    x_train_reduced = pca.fit_transform(x_train)
    x_test_reduced = pca.fit_transform(x_test)
    
    # Run Linear Regression using PCA reduced dataset
    model = ols.fit(x_train_reduced, y_train)
    
    # Record score
    r_scores.append(r2_score(model.predict(x_test_reduced), y_test))
    rmse_scores.append(math.sqrt(mean_squared_error(model.predict(x_test_reduced), y_test)))
```

```python
plt.plot(r_scores)
```

From this plot it seems that we can use PCA to reduce the training set to 15-dimensional space while retaining similar $r^2$ values.

```python
plt.plot(rmse_scores)
```

From the above chart we can see that the RMSE score is also minimized around 15-dimensional space.  Let's use this information to test out a more computationally expensive model using the reduced feature space.

## SVM

```python
pca = PCA(n_components = 15)
x_train_reduced = pca.fit_transform(x_train)
x_test_reduced = pca.fit_transform(x_test)
```

```python
svm = SVR(C=1.0, epsilon=0.2)
svm.fit(x_train_reduced, y_train) 
```

```python
print 'RMSE: ' + str(math.sqrt(mean_squared_error(svm.predict(x_test_reduced), y_test)))
print 'R2: ' + str(r2_score(svm.predict(x_test_reduced), y_test))
```

Unfortunately it doesn't seem like SVM performs well in this setting either...

## Nonlinear Mapping

Perhaps a simple linear model is not equipped to model the dataset well.  To try and mitigate this problem let us apply a nonlinear mapping of each feature ($x_i \rightarrow x_i^k$ for some $1..k$).

```python
nl_x_train = x_train.copy()
nl_x_test = x_test.copy()

for feature in nl_x_train:
    nl_x_train[feature + '_sq'] = nl_x_train[feature] ** 2
    nl_x_test[feature + '_sq'] = nl_x_test[feature] ** 2
```

Now let us experiment with the same algorithms we used before.

```python
# Linear regression
model = ols.fit(nl_x_train, y_train)
print 'RMSE: ' + str(math.sqrt(mean_squared_error(model.predict(nl_x_test), y_test)))
print 'R2: ' + str(r2_score(model.predict(nl_x_test), y_test))
```

```python
# Ridge regression
ridge = linear_model.Ridge(alpha=10)
ridge.fit(nl_x_train, y_train)

print 'RMSE: ' + str(math.sqrt(mean_squared_error(ridge.predict(nl_x_test), y_test)))
print 'R2: ' + str(r2_score(ridge.predict(nl_x_test), y_test))
```

```python
# Lasso
lasso = linear_model.Lasso(alpha=.001)
lasso.fit(nl_x_train, y_train)

print 'RMSE: ' + str(math.sqrt(mean_squared_error(lasso.predict(nl_x_test), y_test)))
print 'R2: ' + str(r2_score(lasso.predict(nl_x_test), y_test))
```

```python
# KNN
knn = KNeighborsRegressor(n_neighbors=2)
knn.fit(nl_x_train, y_train)

print 'RMSE: ' + str(math.sqrt(mean_squared_error(knn.predict(nl_x_test), y_test)))
print 'R2: ' + str(r2_score(knn.predict(nl_x_test), y_test))
```

From these results, it does not seem like the simple nonlinear mapping is worth using for this task.
